```python
df_pop = pd.read_csv("The_World_Bank_Population_growth_(annual_).csv")
df_pop
```

```python
len(list(df_pop.country_name))
```

```python
df_pop.info()
```

```python
df_pop["2023"] = df_pop["2022"].apply(lambda x: x+10)
```

```python
df_pop["country_name"].values[:10]
```

```python
df_pop["country_name"] = df_pop["country_name"].apply(lambda x: x.replace(" ","_"))
```

```python
df_pop.head()
```

```python
df_pop2 = df_pop.set_index("country_name")
df_pop2 = df_pop2.sort_index()
df_pop2.head()
```

```python
df_pop3 = df_pop.set_index(["country_name","country_code"])
df_pop3.head(10)
```

```python
df_pop3.index
```

```python
df_pop3.loc["India"]
```

```python
df_pop3.loc[("India","IND")]
```

```python
df_pop.set_index("country_name").sort_values(["country_code","2021"],ascending=[False,True]).head(10)
```

```python
 df_pop.country_code.value_counts()
```

```python
df_pop4 = df_pop[["country_code","country_name"]].set_index(["country_code","country_name"])
df_pop4.head(10)
```

```python
list_of_years = [str(i) for i in range(1961,2024)]
df_pop4["Avg_population"] = df_pop[list_of_years].mean(axis=1).values
```

```python
df_pop4.sort_values("Avg_population",ascending=False).head(10)
```

```python
df_pop4.sort_values("Avg_population",ascending=False).head(30).plot(kind='barh')
```
